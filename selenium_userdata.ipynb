```python
from selenium import webdriver
from bs4 import BeautifulSoup

import platform
import os 
import time

cwd = os.getcwd()
print(cwd)
os = platform.system()

user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.50 Safari/537.36'    

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
options.add_argument('user-agent={0}'.format(user_agent))

if(os=="Darwin"):
    driver = webdriver.Chrome(cwd + "/chromedriver", options=options)
else:
    driver = webdriver.Chrome(cwd + "\\chromedriver.exe", options=options)

driver.execute_script("return navigator.userAgent")


targetUser = "gauti_98"


driver.get("https://www.zomato.com/" + targetUser+"/reviews")
```

```python
value_find = driver.page_source

pageData = BeautifulSoup(value_find, 'html.parser')
print(pageData.prettify())

noReviews = (pageData.findAll(class_='sc-gGsJSs fVSAtZ')[0].text).split(" ")[0]
print(noReviews)

#noReviews = int(results1.find_all('div', class_='ui label')[0].text)
#print(noReviews)
```

```python
current_count = 0
review_elems = ""

while True:
    more_buttons = driver.find_elements_by_class_name("load-more-feed")

    time.sleep(5)

    for x in range(len(more_buttons)):
        if more_buttons[x].is_displayed():
            driver.execute_script("arguments[0].click();", more_buttons[x])
            time.sleep(5)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    results = soup.findAll(id='tabtop')[0]
    # print(results.prettify())
    review_elems = results.find_all('div', class_='ui segment brtop')
    print(len(review_elems))
            
            
    if(len(review_elems)==current_count):
        break
        
    current_count = len(review_elems)

# html = driver.page_source
driver.quit()
```
